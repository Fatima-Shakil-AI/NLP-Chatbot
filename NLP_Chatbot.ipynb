```python
from transformers import pipeline
chatbot = pipeline("text2text-generation",
                   model="google/flan-t5-base")   # ≈250 MB more but much smarter

def ask_chatbot(q):
    prompt = f"You are a helpful AI assistant.\nUser: {q}\nAssistant:"
    out = chatbot(prompt, max_new_tokens=150, temperature=0.7)
    print("User:", q)
    print("Bot:", out[0]["generated_text"].strip())
```

```python
ask_chatbot("Tell me a short joke about robots.")
```
